```python
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = []
    for page in doc:
        text.append(page.get_text("text"))
    return "\n".join(text)  # Объединяем все страницы в одну строку
```

```python
def parse_habr_pdf(text):
    lines = [line.strip() for line in text.split('\n') if line.strip()]
    
    data = {
        'NameCompany': '',
        'Description': '',
        'Raiting': 0,
        'DataPublish': '',
        'Activity': '',
        'TextArticle': text
    }
    
    # Поиск рейтинга и названия компании
    for i, line in enumerate(lines):
        if re.match(r'^\d+\.?\d*$', line) and i+1 < len(lines) and "Рейтинг" in lines[i+1]:
            try:
                data['Raiting'] = float(line)
            except ValueError:
                data['Raiting'] = 0
            
            if i+2 < len(lines):
                data['NameCompany'] = lines[i+2]
            if i+3 < len(lines):
                data['Activity'] = lines[i+3]
            break
    
    # Поиск даты публикации
    date_patterns = [
        # "10 часов назад", "5 дней назад"
        (r'(\d+\s(?:часов|дней|минут)\sназад)', lambda m: m.group(1)),
        
        # "16 ноября 2008", "29 июля 2011"
        (r'(\d{1,2}\s(?:января|февраля|марта|апреля|мая|июня|июля|августа|сентября|октября|ноября|декабря)\s\d{4})', 
         lambda m: m.group(1)),
        
        # "27.03.2025, 20:55" - берем только дату
        (r'(\d{2}\.\d{2}\.\d{4})(?:, \d{2}:\d{2})?', lambda m: m.group(1)),
        
        # "29 дек 2023 в 18:31" - берем только дату
        (r'(\d{1,2}\s(?:янв|фев|мар|апр|мая|июн|июл|авг|сен|окт|ноя|дек)\s\d{4})(?:\sв\s\d{2}:\d{2})?', 
         lambda m: m.group(1)),
        
        # берем только временной период
        (r'(?:\w+\s)?(\d+\s(?:часов|дней|минут)\sназад)', lambda m: m.group(1)),
        
        (r'(?:\w+\s)?(\d+\s(?:часов|дней|минут)\sназад)', lambda m: m.group(1)),
        
        # "2023-12-29"
        (r'(\d{4}-\d{2}-\d{2})', lambda m: m.group(1))
    ]
    
    for line in lines:
        for pattern, handler in date_patterns:
            match = re.search(pattern, line)
            if match:
                data['DataPublish'] = handler(match)
                break
        if data['DataPublish']:
            break
    return data
```

```python
# Обработка всех PDF файлов
results = []
failed_files = []

for pdf_path in all_pdf:
    try:
        # Извлекаем текст (теперь это строка, а не список)
        text = extract_text_from_pdf(pdf_path)
        
        parsed_data = parse_habr_pdf(text)
        results.append(parsed_data)
        
        print(f"Успешно обработан: {pdf_path.split('\\')[-1]}")
    except Exception as e:
        failed_files.append((pdf_path, str(e)))
        print(f"Ошибка при обработке {pdf_path.split('\\')[-1]}: {str(e)}")
```

```python
df = pd.DataFrame(results)
```
