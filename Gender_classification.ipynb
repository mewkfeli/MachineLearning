### Modelling

```python
from sklearn.model_selection import train_test_split
```

```python
df.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['Favorite Color', 'Favorite Music Genre', 'Favorite Beverage']], 
                                                      df['Gender'], 
                                                      test_size=0.33, 
                                                      random_state=42)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, mean_absolute_error, mean_squared_error, precision_score, recall_score, f1_score, confusion_matrix, r2_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import RocCurveDisplay
```

### KNeighborsClassifier (Классификация ближайших соседей)

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
Y = df['Gender']
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(Y), yticklabels=np.unique(Y))
plt.show()
```

### Gaussian Naive Bayes

```python
gnb = GaussianNB()
```

```python
gnb.fit(X_train, y_train)
```

```python
y_pred_gnb = gnb.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_gnb))
```

```python
conf_matrix = confusion_matrix(y_test, y_pred_gnb)
```

```python
Y = df['Gender']
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(Y), yticklabels=np.unique(Y))
plt.show()
```

### Random Forest Classifier

```python
rfc  = RandomForestClassifier(criterion = 'gini',
                                    n_estimators = 100,
                                    max_depth = 9,
                                    random_state = 42,
                                    n_jobs = -1)
```

```python
rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
rfc_pred_test = rfc.predict(X_test)
display(rfc_pred_test[:20], np.array(y_test[:20]))
```

```python
print(classification_report(y_test, y_pred))
```

```python
conf_matrix = confusion_matrix(y_test, rfc_pred_test)
```

```python
Y = df['Gender']
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(Y), yticklabels=np.unique(Y))
plt.show()
```

### Logistic Regression (Логистическая регрессия)

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
lr = LogisticRegression(random_state = 42)
lr.fit(X_train, y_train)
```

```python
y_pred_lr = lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_lr))
```

```python
conf_matrix = confusion_matrix(y_test, y_pred_lr)
```

```python
Y = df['Gender']
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(Y), yticklabels=np.unique(Y))
plt.show()
```

### Сравнение моделей

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
classifiers =["KNeighborsClassifier", "Gaussian Naive Bayes", "Random Forest Classifier", "Logistic Regression"]

log = pd.DataFrame()

log["Classifier"] = classifiers
log["Accuracy"] = [accuracy_score(y_test, y_pred), accuracy_score(y_test, y_pred_gnb), accuracy_score(y_test, rfc_pred_test), accuracy_score(y_test, y_pred_lr)]

log
```

```python
plt.xlabel('Accuracy')
plt.title('Classifier Accuracy')

sns.set_color_codes("muted")
sns.barplot(x='Accuracy', y='Classifier', data=log.sort_values(by='Accuracy'), color="b");
```

```python
param_grid = {'max_depth': [i for i in range(1,20)]}
grid_search = GridSearchCV(estimator=rfc, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)
print("Best Parameters: ", grid_search.best_params_)

param_grid = {'n_jobs': [i for i in range(-5,5)]}
grid_search = GridSearchCV(estimator=rfc, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)
print("Best Parameters: ", grid_search.best_params_)

param_grid = {'random_state': [i for i in range(20,40)]}
grid_search = GridSearchCV(estimator=rfc, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)
print("Best Parameters: ", grid_search.best_params_)

param_grid = {'n_estimators': [i for i in range(50,100)]}
grid_search = GridSearchCV(estimator=rfc, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)
print("Best Parameters: ", grid_search.best_params_)
```
